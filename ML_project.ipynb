Training model

```python
from sklearn.preprocessing import OrdinalEncoder,StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
x=df.drop(columns=['selling_price'])
y=df['selling_price']
```

```python
ordinal_encoder=OrdinalEncoder()
x['owner']=ordinal_encoder.fit_transform(x[['owner']])
```

```python
x_encoded=pd.get_dummies(x,columns=['name','fuel','seller_type','transmission'],drop_first=True)
```

```python
# Select boolean columns (usually dummies)
bool_cols = x_encoded.select_dtypes(include='bool').columns

# Convert only those to int
x_encoded[bool_cols] = x_encoded[bool_cols].astype(int)
x_encoded
```

```python
x_train,x_test,y_train,y_test=train_test_split(x_encoded,y,test_size=0.3,random_state=42)
```

```python
scaler = StandardScaler()
num_cols = ['age', 'km_driven', 'seats', 'max_power (in bph)', 'Mileage', 'Engine (CC)']
x_train[num_cols] = scaler.fit_transform(x_train[num_cols])
x_test[num_cols] = scaler.transform(x_test[num_cols])
x_train
# scaler=StandardScaler()
# x=scaler.fit_transform(x_encoded)
# x
```

```python
model=LinearRegression()
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
y_pred
```

```python
y_test
```

```python
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import cross_val_score
from sklearn.metrics import r2_score
```

```python
mse=mean_squared_error(y_test,y_pred)
rmse=np.sqrt(mse)
scores = cross_val_score(model, x_encoded, y, cv=5, scoring='r2')

print(f"MSE (Computer's Score): {mse:.2f}")
print(f"RMSE (Average Error in Price): {rmse:.2f}")
print(f"Average R2 across 5 tests: {scores.mean():.2f}")

# Calculate R2
score = r2_score(y_test, y_pred)
print(f"Model Accuracy (R2 Score): {score * 100:.2f}%")
```
